```python
class Dog:
    def __init__(self, name):
        self.name = name
    def bark(self):
        if self.name == 'Bholu':
            print(f"{self.name} barks bhow bhow!!!")
        else: 
            print(f"{self.name} barks lau lau!!!")

dog1 = Dog('Bholu')
dog1.bark()

dog2 = Dog('Lalu')
dog2.bark()

```

```python
class Shape:
    side = 3
    def __init__(self, side):  # Instance/object method
        self.side = side
    def tri1(self):
        print(f"Triangle has {self.side} sides")
    
    @classmethod # Class method
    def tri2(cls):
        print(f"Triangle has {cls.side} sides")
    
    @staticmethod # Static method
    def tri3(side):
        print(f"Triangle has {side} sides")

t1 = Shape(3)
t1.tri1()

Shape.tri2()

Shape.tri3(3)
```

```python
class Bank:
    bname = "SBI"
    branch = "Hyderabad"
    ifsc = "SBIN000123"
    helpline = "1800-11-2211"

    def __init__(self, name, age, phone, pan, balance):
        self.name = name
        self.age = age
        self.phone = phone
        self.pan = pan
        self.balance= balance

    def display(self):
        print(self.name, self.phone, self.balance)

    @classmethod
    def changeName(cls):
        cls.bname = "PNB"
    
    @classmethod
    def printName(cls):
        print(cls.bname)

Bank.printName()
Bank.changeName()
Bank.printName()
```
