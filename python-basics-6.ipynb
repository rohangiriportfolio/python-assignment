```python
# Encapsulation  # Private/Mangling members,variable, method

# Protected members

class A:
    def __init__(self, a):
        self.a = a 
    def _printA(self):
        print(self.a)

class B(A):
    def __init__(self, a):
        super().__init__(a)
    def printA(self):
        super()._printA()

b = B(1)
b.printA()
b._printA() # Not recomended in the production as protected memeber can't be accessed outside the class except subclass
```

```python
# Private members

class A:
    def __init__(self):
        pass

    def test(self, b):
        self.b = b

    def setter(self, a):  # Setter to set the value
        self.__a = a

    def getter(self): # Getter to get the value
        print("This is private var a:", self.__a)


a = A()
a.setter(10)
a.getter()
a.test(20)

print("This is public var b:", a.b)
# a.__a  # Will give error on direct call as it is private
```

```python
from abc import ABC, abstractmethod

class Bank(ABC):
    @abstractmethod  # Abstract Method
    def rate(self):
        pass

class SBI(Bank):
    def rate(self, a):
        print(f"The interest is {a}%")

bank1 = SBI()
bank1.rate(7)

# bank1 = Bank()  # Can't instantiate abstract class Bank without an implementation for abstract method 'rate'
```

```python
from abc import ABC, abstractmethod

class BankAccount(ABC):
    def __init__(self, cus_name, dob):
        self.cus_name = cus_name
        self.dob = dob

    @abstractmethod
    def deposit(self, amount):
        pass
    @abstractmethod
    def withdraw(self, amount):
        pass

class SavingAccount(BankAccount):
    def __init__(self, cus_name, dob, min_bal):
        super().__init__(cus_name, dob)
        self.min_bal = min_bal

    def deposit(self, amount):
        self.min_bal += amount

    def withdraw(self, amount):
        if amount > self.min_bal:
            print("Insufficient Balance!")
        else:
            self.min_bal -= amount
            print("Money withdraw successful using savings account!")
    def bal_amount(self):
        print("Current balance:", self.min_bal)

class CurrentAccount(BankAccount):
    od_limit = 5000
    def __init__(self, cus_name, dob, bal):
        super().__init__(cus_name, dob)
        self.bal = bal - self.od_limit

    def deposit(self, amount):
        self.bal += amount

    def withdraw(self, amount):
        if amount > self.bal + self.od_limit:
            print("Insufficient Balance!")
        else:
            self.bal -= amount
            print("Money withdraw successful using current account!")

        

    def bal_amount(self):
        if self.bal < 0:
            print(f"{self.bal * -1} need to pe pay") 
        else: 
            print("Current balance:", self.bal)


s = SavingAccount('Jenny', '2000', 1000)
s.deposit(1000)
s.bal_amount()
s.withdraw(500)
s.bal_amount()

c = CurrentAccount('Sam', '2001', 100)
c.withdraw(5000)
c.bal_amount()
c.deposit(1000)
c.bal_amount()
c.withdraw(2100)
c.bal_amount()

c.deposit(10000)
c.bal_amount()



```
