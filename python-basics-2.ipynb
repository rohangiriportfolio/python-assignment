```python
def evenOdd(n):
    n1=n
    odd=0
    even=0
    while n1!=0:
        rem = n1%10
        if rem % 2 != 0:
            odd+=1
        else:
            even+=1
        n1//=10
    print("Odd: ",odd," Even: ",even)

n = eval(input("Enter a no: "))
evenOdd(n)
```

```python
#1. Positional argument

def test(a, b):  #a is formal argument
    print(a, b)
test(10, 20) #v is actual argument
```

```python
#2. Default argument

def test(a, b=20):
    print(a, b)
test(10)
```

```python
#3. Keyword argument

def test(a, b):
    print(a, b)
test(b=20, a=10)
```

```python
#4. Variable length positional argument

def add(*args):
    print(sum(args))
add(10, 20)
```

```python
#5. Variable length keyword argument

def test(**kwargs):
    for key, val in kwargs.items():
        print(key, val)
test(a=10, b=20, c=30)
```

```python
#Logging instead of print()
import logging

logging.basicConfig(
    filename="app.log",
    level=logging.INFO,
    format="%(asctime)s - %(lineno)d - %(levelname)s - %(message)s",
    filemode="a"
)

def division(a: int, b: int) -> int:
    if b!=0:
        logging.info(f'Dividing {a} by {b} will get {a//b}')
    else:
        logging.error("Dividing by zero error!!")
a = eval(input("Enter 1st no: "))
b = eval(input("Enter 2nd no: "))
division(a, b)
```

```python
import logging

logging.basicConfig(
    filename="vowel.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def countVow(s: str) -> int:
    """
    Count the number of vowels in a given string.
    Parameters:
        s (str): Input string
    Returns:
        int: Total number of vowels in the string.
    """
    v = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    c = 0
    for i in s:
        if i in v:
            c+=1
    return c
    
s = input()
c=countVow(s)
logging.info(f'No of vowels: {c}')
```

```python
import logging

logging.basicConfig(
    filename="freq.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
) 

def freq(s: str) -> dict:
    """
    Count the frequency of each letter.
    Parameters:
        s (str): Input string
    Returns:
        dict: Frequency of each letter.
    """
    d={}
    for i in s:
        if i in d:
            d[i] = d[i] + 1
        else:
            d[i] = 1
    return d
s = input()
logging.info(f"Frequency of letters: {freq(s)}")
```

```python
import logging

logging.basicConfig(
    filename="secondLarge.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
) 

def secondLarge(l: list) -> int:
    """
    Find the second largest element.
    Parameters:
        l (list): Input list
    Returns:
        b (int): Second largest element.
    """
    a = float('-inf')
    b = float('-inf')
    for i in l:
        if i>a:
            b=a
            a=i
        if i>b and i!=a:
            b=i
    return b

l = [1,2,3,4,5,6]
logging.info(f"Second largest element: {secondLarge(l)}")
```

```python
import logging

logging.basicConfig(
    filename="sort.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
) 

def sorting(l: list) -> None:
    """
    Find the second largest element.
    Parameters:
        l (list): Input list
    Returns:
        None: Only prints is sorted or not.
    """
    l1=l
    l2=sorted(l1)
    if(l2==l):
        logging.info("Sorted in ascending order!")
    else:
        logging.info("Not sorted!")
l=[1,2,3,4]
sorting(l)
```
