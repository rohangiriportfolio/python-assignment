```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, roll, addr):
        super().__init__(name, age)
        self.roll = roll
        self.addr = addr

class Academic:
    def __init__(self, subject, marks):
        self.subject = subject
        self.marks = marks

class Sports:
    def __init__(self, sport_name, level):
        self.sport_name = sport_name
        self.level = level

class AlrounderStudent(Student, Academic, Sports):
    def __init__(self, name, age, roll, addr, is_aca, is_spo, subject=None, marks=None, sport_name=None, level=None):
        super().__init__(name, age, roll, addr)
        self.is_aca = is_aca
        self.is_spo = is_spo
        if self.is_aca:
            Academic.__init__(self, subject, marks)
        if self.is_spo:
            Sports.__init__(self, sport_name, level)

    def display(self):
        print(self.name, self.age, self.roll, self.addr)

        if self.is_aca:
            print(self.subject, self.marks)
        if self.is_spo:
            print(self.sport_name, self.level)



s1 = AlrounderStudent('Kelly', 25, 123, 'Scotland', True, False, 'Math', 92)
s1.display()
# print(s1.__dict__)
```

```python
def pol(a, b):  # Method overloading
    a, b = b, a
    print(a, b)

monkeyPatching = pol   # Monkey Patching

def pol():
    print("Abc")

pol()
monkeyPatching(10, 20)
```

```python
class A:  # Method Overriding
    def printA(a):
        print(a)
class B(A):
    def printA(a, b):
        print(a, b)

a = A.printA('Abc')
b = B.printA(20, 10)
```

```python
class Add:
    def __init__(self, a):
        self.a = a
    def __add__(self, other):
        return self.a + other.a
    def __sub__(self, other):
        return self.a - other.a

a = Add(10)

b = Add(20)

print(a+b)
```

```python
class Vehicle:
    def __init__(self, v_id, brand):
        self.v_id = v_id
        self.brand = brand

    def cal_rent(self, t, x):
        if t == 'Car':
            print("Rent:", x*1000)
        elif t == 'Bike':
            print("Rent:", x*500) 
        else:
            print("Rent:", x*100)

class Car(Vehicle):
    def __init__(self, v_id, brand):
        super().__init__(v_id, brand)

    def cal_rent(self, x):
        print("Rent on basis of days")
        super().cal_rent(self.__class__.__name__, x)


class Bike(Vehicle):
    def __init__(self, v_id, brand):
        super().__init__(v_id, brand)

    def cal_rent(self, x):
        print("Rent on basis of hrs")
        super().cal_rent(self.__class__.__name__, x)

class Truck(Vehicle):
    def __init__(self, v_id, brand):
        super().__init__(v_id, brand)

    def cal_rent(self, x):
        print("Rent on basis of distance travelled")
        super().cal_rent(self.__class__.__name__, x)
    
car = Car(1, 'ab')
car.cal_rent(10)
```
