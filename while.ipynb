```python
n = eval(input("Enter a number: "))
n1 = n
n2 = 0
while n1>0:
    rem = n1%10
    n2 = n2*10 + rem
    n1 = n1//10
if(n2>n):
    print("Reversed no is greater!")
else:
    print("Reversed no is smaller!")
# print(n2)
```

```python
n = eval(input("Enter a no: "))
n1=n
odd=0
even=0
while n1!=0:
    rem = n1%10
    if rem % 2 != 0:
        odd+=1
    else:
        even+=1
    n1//=10
print("Odd: ",odd," Even: ",even)
```

```python
n = eval(input("Enter a number: "))
n1 = n
n2 = 0
while n1>0:
    rem = n1%10
    n2 = n2*10 + rem
    n1 = n1//10
if n == n2:
    print("Palindrome")
else:
    print("Not Palindrome")
```
