```python
# Handling binary files

with open('a.txt') as f:
    w = f.read().split()
    print(len(w))
    print(w)
```

```python
with open('a.txt') as f:
    w = f.read().split(',')  # On basis of comma
    print(len(w))
    print(w)
```

```python
with open('a.txt') as f:
    w = set(f.read().split())  # Unique Words
    print(len(w))
    print(w)
```

```python
with open('a.txt') as f:
    w = f.read().split()
    d = {}
    for i in w:
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    print(d)
```

```python
# Handling JSON files

data = {
    "name": "Rohan",
    "age": 21,
    "dept": "CSE"
}

import json
with open('data.json', 'r+') as f:
    json.dump(data, f)
    f.seek(0)
    print(json.load(f))
```

```python
import json
with open('data.json', 'r+') as f:
    json.dump(data, f)
    f.seek(0)
    d = json.load(f)
    print(len(d))
```

```python
# Handling  CSV files

import csv

with open('data.csv', 'w', newline="") as f:
    w = csv.writer(f)
    w.writerow(['Name', 'Age', 'Dept'])
    w.writerow(['Rohan', 21, 'CSE'])
```

```python
import csv

with open('data.csv', 'w', newline="") as f:
    fns = ['Name', 'Age', 'Dept']
    w = csv.DictWriter(f, fieldnames=fns) # For writing as objects or dictionaries
    w.writeheader()
    w.writerow({'Name': 'Rohan', 'Age': 21, 'Dept': 'CSE'})
```

```python
with open('data.csv', 'r') as f:  # DictReader
    r = csv.DictReader(f)
    print(r)
    for i in r:
        print(i)
```

```python
# x = np.arange(1, 11, .1)
x = np.arange(1, 11)
y = np.sin(x)
z = np.cos(x)
print(x, y, z, sep="\n")
```

```python
plt.plot(x, y)
```

```python
plt.figure(figsize=[5, 5])
plt.plot(x, y, label = 'Sin')
plt.plot(x, z, label = 'Cos')
plt.legend()
plt.title("Sin Vs Cos")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.show()
```

```python
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1) # Row No 1, Col No 1, Plot No 1
plt.plot(x, y, label = 'Sin')
plt.plot(x, z, label = 'Cos')
plt.legend()
plt.title("Sin Vs Cos")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")


plt.subplot(1, 2, 2)
plt.plot(x, y, label = 'Sin')
plt.plot(x, z, label = 'Cos')
plt.legend()
plt.title("Sin Vs Cos")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.show()
```

```python
x = np.arange(1, 21, .1)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(7, 7))

plt.subplot(3, 3, 1) # Row No 1, Col No 1, Plot No 1
plt.plot(x, y, label = 'Sin')
plt.plot(x, z, label = 'Cos')
plt.legend()
plt.title("Sin Vs Cos")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")


plt.subplot(3, 3, 5)
plt.plot(x, y, label = 'Sin')
plt.plot(x, z, label = 'Cos')
plt.legend()
plt.title("Sin Vs Cos")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")

plt.subplot(3, 3, 9)
plt.plot(x, y, label = 'Sin')
plt.plot(x, z, label = 'Cos')
plt.legend()
plt.title("Sin Vs Cos")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.show()
```

```python
l = [1, 5, 9]

plt.figure(figsize=(7, 7))
for i in l:
    x = np.arange(1, 21, .1)
    y = np.sin(x)
    z = np.cos(x)


    plt.subplot(3, 3, i)
    plt.plot(x, y, label = 'Sin')
    plt.plot(x, z, label = 'Cos')
    plt.legend(loc='upper right')
    plt.title("Sin Vs Cos")
    plt.xlabel("Data Points")
    plt.ylabel("Associate Values")
    plt.subplots_adjust(wspace=0.5)  # Space between two plots
    plt.grid()
```
