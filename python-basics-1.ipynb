```python
s = eval(input())
last = s[-1]
if last.isdigit():
    print("Ends with digit")
elif last.isalpha():
    print("Ends with character")
else:
    print("Ends with special symbol")
```

```python
s = eval(input())
if s and s[0].isupper():
    print("Starts with uppercase")
elif s and s[-1].islower():
    print("Ends with lowercase")
else:
    print("Other case")
```

```python
lst = eval(input())
n = len(lst)
first = sum(lst[:n//2])
second = sum(lst[n//2:])
if first > second:
    print("First half greater")
elif first == second:
    print("Equal halves")
else:
    print("Second half greater")
```

```python
s = eval(input())
if len(s) >= 5:
    mid = s[len(s)//2]
    if mid in 'aeiouAEIOU':
        print("Middle character is a vowel")
else:
    print("String too short")
```
