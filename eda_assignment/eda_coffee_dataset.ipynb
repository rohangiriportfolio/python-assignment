****14) What is the total coffee production for each year?****

```python
for y in df['Year'].unique():
    print(y, df[df['Year'] == y]['Total Production'].sum())
```

```python
yearly_production = df.groupby('Year')['Total Production'].sum()

print(yearly_production)
```

```python
import matplotlib.pyplot as plt
plt.plot(yearly_production.index, yearly_production.values)
plt.xlabel("Year")
plt.ylabel("Total Production")
plt.title("Year-wise Coffee Production")
plt.show()
```

****15) Which are the top 5 countries based on total production?****

```python
top5 = df.groupby('Country')['Total Production'].sum().sort_values(ascending=False).head(5)

top5
```

```python
plt.barh(top5.index, top5.values, color = ['red', 'blue', 'green', 'yellow', 'orange'])
plt.xlabel("Total Production")
plt.ylabel("Country")
plt.title("Top 5 Coffee Producing Countries")
plt.show()
```

****16) Which country has the highest average export price?****

```python
avg_price = df.groupby('Country')['Export Price'].mean()
avg_price.sort_values(ascending=False).head(1)
```

****17) Does domestic consumption exceed total production for any country?****

```python
country_totals = df.groupby('Country')[['Domestic Consumption', 'Total Production']].sum()
country_totals[country_totals['Domestic Consumption'] > country_totals['Total Production']]
```

****18) Which country shows the highest growth in total production over the years?****

```python
max_prod = df.groupby('Country')['Total Production'].max()
min_prod = df.groupby('Country')['Total Production'].min()
growth = max_prod - min_prod
highest_growth = growth[growth == growth.max()]
print(highest_growth)
```

```python
plt.barh(growth.index, growth.values, color = ['red', 'green', 'blue'])
plt.xlabel("Production Growth")
plt.ylabel("Country")
plt.title("Production Growth by Country")
plt.show()
```

****19) Is there any relationship between export price and total production?****

```python
df['Export Price'].corr(df['Total Production'])
```

```python
plt.scatter(df['Total Production'], df['Export Price'])
plt.xlabel("Total Production")
plt.ylabel("Export Price")
plt.title("Export Price vs Total Production")
plt.show()
```
